```python
#Import Libraries
import pandas as pd
import numpy as np
#SkLearn
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
#Test Our Model
from sklearn.metrics import accuracy_score

#Read Data
data = pd.read_csv('diabetes.csv')

y = np.array(data['Outcome'])
x = np.array(data.drop(['Outcome'],axis=1))
star = StandardScaler()
X = star.fit_transform(x)
x_train , x_test , y_train , y_test = train_test_split(X,y,test_size=0.1,random_state=1)
model = KNeighborsClassifier(n_neighbors=7,metric='euclidean')
model.fit(x_train,y_train)

pre_train = model.predict(x_train)
pre_test = model.predict(x_test)

#Acc 1
acc_test = model.score(x_test,y_test)
print("TestACC: ", acc_test)
acc_train = model.score(x_train,y_train)
print("TrainACC: ", acc_train)
#Or
# print("TestACC: ", accuracy_score(y_test,pre_test))
# print("TrainACC: ", accuracy_score(y_train,pre_train))



for i in range(len(pre_test)):
    if pre_test[i] != y_test[i]:
        print()
        print(f'No {i} Predicted : {pre_test[i]}')
        print(f'No {i} True Data : {y_test[i]}')
```
