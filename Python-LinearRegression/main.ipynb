```python
#Import Modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
from sklearn import linear_model
#Read Data
df = pd.read_csv('student-por.csv',sep=';')
#Replace Sex Values
df['sex'] = df['sex'].apply(lambda x: 1 if x == "M" else 0)
#Percentages Of A Value
#Formula
# # percernt = df.sex.value_counts(normalize=True).mul(100).round(1)
# counts = df.sex.value_counts()
# percs = df.sex.value_counts(normalize=True).mul(100).round(1)
# result = pd.concat([counts,percs], axis=1, keys=['count', 'percentage'])



###----------------MAIN----------------###

df = df[["sex","age","absences","freetime","G1","G2","G3"]]
predict = "G3"
X = np.array(df.drop([predict],axis=1))
y = np.array(df[predict])
x_train , x_test , y_train , y_test = train_test_split(X,y,test_size=0.2)
my_model = linear_model.LinearRegression()
my_model.fit(x_train,y_train)
acc = my_model.score(x_test,y_test)
print(f"acc => {acc}")
print(f"Coef => {my_model.coef_}")
print(f"intercept => {my_model.intercept_}")
print()

pre = my_model.predict(x_test)
for i in range(len(pre)):
    print(f"Data => {y_test[i]} Predict => {round(pre[i])}")

###----------------Test Model----------------###
Mcoef = my_model.coef_
Minter = my_model.intercept_ 
sum = 0
for i in range(len(my_model.coef_)):
    formula = Mcoef[i] * x_test[2][i]
    sum += formula
sum += Minter
Mp = float(my_model.predict([x_test[2]]))
print(f' Machine Predict : {round(Mp,1)} What I Found! : {round(sum,1)}')
```
